```python
#Okay but now we want to see where the birds are
#A heat map easily tells us the density of birds. 

birddens = zip(df['lat'], df['lng'], df['howMany'])
list_of_zip = list(birddens)
list_of_zip[0:10]
```

We may get an error thrown for dupe lat/lng...but maybe not bc heat map solution

```python
hm_layer = HeatMap(list_of_zip,
                   min_opacity = 0.2,
                   radius = 10
                   blur = 6,
                  )
```

```python
folium_hmap.add_child(hm_layer)
folium_hmap
```
